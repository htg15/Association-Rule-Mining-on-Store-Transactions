Retail Store Association Rule Mining. This data set contains 7500 transactions over the course of a week at a French retail store.

Write a program to find interesting association rules in this data set.

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
import pandas as pd
import numpy as np
```

```python
#Reading the CSV to a dataframe

store_data=pd.read_csv('store_data.csv')
```

```python
#Checking data

store_data.head()
```

```python
#Converting dataframe values of floats to string to make sure and convert it to list of lists

store_data_ls = []
for i in range(0,7500):
    store_data_ls.append([str(store_data.values[i,j]) for j in range(0, 20)])

store_data_ls = np.array(store_data_ls)
store_data_ls.shape
```

```python
te=TransactionEncoder()
te_arr=te.fit_transform(store_data_ls)
te_arr
```

```python
#verify column names
te.columns_
```

```python
store_df=pd.DataFrame(te_arr,columns=te.columns_)
store_df.head(),store_df.shape
```

```python
#Ignoring nans
store_dff=store_df.drop('nan',axis=1)
```

```python
frequent_itemsets=fpgrowth(store_dff,min_support=0.1,use_colnames=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets = fpgrowth(store_dff, min_support = 0.05, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets_2 = apriori(store_dff, min_support=0.1, use_colnames = True)
frequent_itemsets_2
association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.2)
```

```python
rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1.2)
rules
```

The top scoring rule with highest confidence is Spaghetti -> Mineral Water, and Choclate -> Mineral Water. We can see that these 2 rules have highest lift along with a good support values. 

These rules signifies that every transaction having Spaghetti or Choclate as an antecedent will have a consequent of mineral water in their basket. Having Spaghetti or Choclate in the basket means that the next item the person purchases is most likely to be mineral water.

Randomly split your records into two sets with roughly 50% of data each. Now use the Apriori algorithm to determine rules on both of these sets. Do you find similar rules on both sets? What does the similary/the differences indicate?

```python
#import library to split the data into 2 fractions at random

from sklearn.model_selection import train_test_split
store_1,store_2=train_test_split(store_data,test_size=0.5,random_state=42)
```

```python


store_1_ls = []
store_2_ls = []
for i in range(0, 3750):
    store_1_ls.append([str(store_1.values[i,j]) for j in range(0, 20)])
    store_2_ls.append([str(store_2.values[i,j]) for j in range(0, 20)])

# conveting it into an numpy array
store_1_ls = np.array(store_1_ls)
store_2_ls = np.array(store_2_ls)
# checking the shape of the array
store_1_ls.shape,store_2_ls.shape
```

```python
te1=TransactionEncoder()
te_arr_1 = te1.fit_transform(store_1_ls)
te2=TransactionEncoder()
te_arr_2 = te2.fit_transform(store_2_ls)
```

```python
final_data_1 = pd.DataFrame(te_arr_1, columns = te1.columns_)
final_data_1
```

```python
final_data_2 = pd.DataFrame(te_arr_2, columns = te2.columns_)
final_data_2
```

```python
#Ignoring nans

final_data_1=final_data_1.drop('nan',axis=1)
final_data_2=final_data_2.drop('nan',axis=1)
```

```python
# Applying Apriori on itemset 1

frequent_itemsets_1 = apriori(final_data_1, min_support=0.1, use_colnames = True)
frequent_itemsets_1
```

```python
# Applying Apriori on itemset 2

frequent_itemsets_2 = apriori(final_data_2, min_support=0.1, use_colnames = True)
frequent_itemsets_2
```

We can notice that the frequent itemsets from both halves of the parent data are very similar and only have a minimal difference in the support and the itemsets.

The support difference between the 2 frequent itemsets is very similar and is differed by 0.02 atmost.
